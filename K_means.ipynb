```python
def kmeans(X, K):
    centorids = X[np.random.choice(len(X), K)]
    centorids_iter = np.empty((K, X.shape[1]))
    while np.not_equal(centorids, centorids_iter).all():
        centorids_iter = centorids
#         print("this is centorids:", centorids)
        distance = [[np.dot(X_i-centorids_i, X_i-centorids_i) for centorids_i in centorids] for X_i in X]
#         print("this is distance:", distance)
        kinds = np.argmin(distance, axis=1)
#         print("this is kinds:", kinds)
        centorids = [np.mean(X[kinds == i], axis=0) for i in range(K)]
        plt.plot(centorids)
    return centorids
```
